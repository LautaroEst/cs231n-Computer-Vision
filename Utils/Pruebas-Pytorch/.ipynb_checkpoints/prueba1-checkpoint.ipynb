Lo primero que hay que hacer es cargar el dataset. En este caso, vamos a cargar el dataset MNIST en una variable `mnist_train` usando la función `datasets` del módulo [torchvision](https://pytorch.org/docs/stable/torchvision/index.html).

```python
# Obtenemos el dataset
from torchvision.transforms import ToTensor, Compose
mnist_train = dset.MNIST('./MINST/', train=True, download=True, transform=ToTensor())

# Información sobre el dataset:
print('Dataset lenght: ', len(mnist_train))
img, num = mnist_train[0]
print('Primera muestra del dataset: ', num)
fig, ax = plt.subplots()
ax.imshow(img[0,:,:])
```

Ahora, para el paso de optimización, no voy a usar todos los datos a la vez. Necesito pasarle a la función de costo un batch de datos del dataset. Para eso, voy a usar un objeto `Dataloader` que recibe instrucciones de cómo samplear el dataset y devuelve el verdadero dataset que se usa en cada paso de optimización.

```python
# Defino un sampler:

class MySampler(Sampler):
    def __init__(self,data_source):
        self.data_source = data_source
        self.num_samples = len(data_source)
        
    def __iter__(self):
        return iter(list(np.random.permutation(self.num_samples)))
    
    def __len__(self):
        return self.num_samples
    
NUM_TRAIN = len(mnist_train)
mnist_train_dataloader = DataLoader(mnist_train, batch_size=64, sampler=MySampler(range(NUM_TRAIN)))
for t, (x,y) in enumerate(mnist_train_dataloader):
    print('Número de muestra: ',t)
    print('Muestra de entrada: ', x.size())
    print('Muestra de salida: ',y.size())
    print()
```
