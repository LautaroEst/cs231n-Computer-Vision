```python
import torch
from pytorch_utils import *
```

```python
NUM_TRAIN = 50000
NUM_VAL = 10000
train_dataloader, val_dataloader, test_dataloader = GetMNISTDataLoaders(NUM_TRAIN, NUM_VAL, batch_size=64)

for t, (x,y) in enumerate(train_dataloader):
    print('Número de muestra: ',t)
    print('Muestra de entrada: ', x.size())
    print('Muestra de salida: ',y.size())
    print()
```

```python
import torch.nn as nn

class SoftmaxClassifier(nn.Module):
    
    def __init__(self,input_shape,output_shape):
        super().__init__()
        H, W = input_shape
        self.linear_layer = nn.Linear(H * W, output_shape)
        self.softmax_layer = nn.LogSoftmax(dim=0)
        self.loss_function = nn.NLLLoss()
        
    def forward(self,x):
        x = x.view(x.shape[0],-1)
        scores = self.softmax_layer(self.linear_layer(x))
        return scores
    
    def loss(self,scores,target):
        return self.loss_function(scores,target)
    
    
input_shape = (28, 28)
output_shape = 10
model = SoftmaxClassifier(input_shape, output_shape)
```

```python
import torch.optim as optim

def CheckAccuracy(loader, model):
    if loader.dataset.train:
        print('Checking accuracy on validation set')
    else:
        print('Checking accuracy on test set')   
    num_correct = 0
    num_samples = 0
    model.eval()  # set model to evaluation mode
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device=device, dtype=torch.float32)  # move to device, e.g. GPU
            y = y.to(device=device, dtype=torch.long)
            scores = model(x)
            _, preds = scores.max(1)
            num_correct += (preds == y).sum()
            num_samples += preds.size(0)
        acc = float(num_correct) / num_samples
        print('Got %d / %d correct (%.2f)' % (num_correct, num_samples, 100 * acc))


epochs = 1 # Cantidad de epochs
device = 'cpu' # CPU / GPU ?
learning_rate = 1e-2 # Tasa de aprendizaje
print_every = 10

# Entrenamiento:
model = model.to(device=device)
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
for e in range(epochs):
    for t, (x,y) in enumerate(train_dataloader):
        x = x.to(device=device, dtype=torch.float32)
        y = y.to(device=device, dtype=torch.long)
        
        # Forward pass
        scores = model(x)
        
        # Backward pass
        loss = model.loss(scores,y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if t % print_every == 0:
            print('Iteration %d, loss = %.4f' % (t, loss.item()))
            CheckAccuracy(val_dataloader, model)
            print()

```
